```python
import random
```

```python
A=[]
B=[]
for i in range(100):
  pa=random.uniform(0,1)
  pb=random.uniform(0,1)
  pm=random.uniform(0,1)
  ta=pa+4
  tb=pb*(9*pm+5)+1
  A.append(ta)
  B.append(tb)

print(A)
```

```python
import matplotlib.pyplot as plt

plt.plot(range(100),A)
plt.plot(range(100),B)
```

```python
A=[]
B=[]
for i in range(100):
  pa=random.uniform(0.65,0.75)
  pb=random.uniform(0.25,0.35)
  pm=random.uniform(0.15,0.25)
  ta=pa+4
  tb=pb*(9*pm+5)+1
  A.append(ta)
  B.append(tb)

plt.plot(range(100), A, label= "Maquina A")
plt.plot(range(100), B, label= "Maquina B")
plt.legend()
```

```python
n=10
A=[]
B=[]
for i in range(n):
  pa=random.uniform(0.65,0.75)
  pb=random.uniform(0.25,0.35)
  pm=random.uniform(0.15,0.25)
  ta=pa+4
  tb=pb*(9*pm+5)+1
  A.append(ta)
  B.append(tb)

plt.plot(range(n), A, label= "Maquina A")
plt.plot(range(n), B, label= "Maquina B")
plt.legend()
```
